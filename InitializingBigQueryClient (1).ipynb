```python
from google.cloud import bigquery

# Initialize BigQuery client
client = bigquery.Client()

print("BigQuery client initialized successfully!")
```

```python
query = """
SELECT *
FROM `exam-dec-2024.examdataset.examtable`
LIMIT 1;
"""
query_job = client.query(query)
results = query_job.result()
for row in results:
    print(row)
```

```python
#Daily Vaccinations Trend for United States
query = """
SELECT date, daily_vaccinations
FROM `exam-dec-2024.examdataset.examtable`
WHERE country = 'United States'
ORDER BY date;
"""
query_job = client.query(query)
results = query_job.result()

for row in results:
    print(f"Date: {row.date}, Daily Vaccinations: {row.daily_vaccinations}")
```
