```python
from google.cloud import bigquery

# Initialize BigQuery client
client = bigquery.Client()

print("BigQuery client initialized successfully!")
```

```python
query = """
SELECT
    country,
    MAX(total_vaccinations) AS max_total_vaccinations
FROM
    `examdataset.examtable`
GROUP BY
    country
ORDER BY
    max_total_vaccinations DESC
LIMIT 1;
"""
query_job = client.query(query)
results = query_job.result()
for row in results:
    print(row)
```

```python
query = """
SELECT
    date,
    SUM(daily_vaccinations) AS total_daily_vaccinations
FROM
    `examdataset.examtable`
GROUP BY
    date
ORDER BY
    date;
"""
query_job = client.query(query)
results = query_job.result()
for row in results:
    print(row)
```

```python
query = """
SELECT
    vaccines,
    COUNT(DISTINCT country) AS countries_using
FROM
        `examdataset.examtable`
GROUP BY
    vaccines
ORDER BY
    countries_using DESC;
"""
query_job = client.query(query)
results = query_job.result()
for row in results:
    print(row)
```

```python
query = """
SELECT
    country,
    MAX(people_fully_vaccinated_per_hundred) AS max_fully_vaccinated_percent
FROM
    `examdataset.examtable`
GROUP BY
    country
ORDER BY
    max_fully_vaccinated_percent DESC;
"""
query_job = client.query(query)
results = query_job.result()
for row in results:
    print(row)
```

```python
query = """
SELECT
    country,
    AVG(daily_vaccinations) AS avg_daily_vaccinations,
    MAX(people_fully_vaccinated_per_hundred) AS max_fully_vaccinated_percent
FROM
    `examdataset.examtable`
GROUP BY
    country
ORDER BY
    avg_daily_vaccinations DESC;
"""
query_job = client.query(query)
results = query_job.result()
for row in results:
    print(row)
```

```python
# Configure export job

destination_uri = "gs://ccbd-exam-2024-alghasi/query_results_*.csv"
extract_job_config = bigquery.ExtractJobConfig(
    destination_format=bigquery.DestinationFormat.CSV,
    field_delimiter=",",  # Delimiter for CSV
    print_header=True     # Include headers in export
)

# Reference the query results table
table_id = "exam-dec-2024.examdataset.examtable"

# Create the extract job
extract_job = client.extract_table(
    source=table_id,  # The table with query results
    destination_uris=destination_uri,  # Where to save results in GCS
    job_config=extract_job_config
)

# Wait for the extract job to complete
extract_job.result()
print(f"Exported query results to {destination_uri}")
```

```python
from google.cloud import bigquery

client = bigquery.Client()

# List of queries to execute
queries = [
    """
SELECT
    country,
    MAX(total_vaccinations) AS max_total_vaccinations
FROM
    `examdataset.examtable`
GROUP BY
    country
ORDER BY
    max_total_vaccinations DESC
LIMIT 1;
"""
    ,
    """
SELECT
    date,
    SUM(daily_vaccinations) AS total_daily_vaccinations
FROM
    `examdataset.examtable`
GROUP BY
    date
ORDER BY
    date;
"""
    ,
 """
SELECT
    vaccines,
    COUNT(DISTINCT country) AS countries_using
FROM
        `examdataset.examtable`
GROUP BY
    vaccines
ORDER BY
    countries_using DESC;
"""
,
"""
SELECT
    country,
    MAX(people_fully_vaccinated_per_hundred) AS max_fully_vaccinated_percent
FROM
    `examdataset.examtable`
GROUP BY
    country
ORDER BY
    max_fully_vaccinated_percent DESC;
"""
,
 """
SELECT
    country,
    AVG(daily_vaccinations) AS avg_daily_vaccinations,
    MAX(people_fully_vaccinated_per_hundred) AS max_fully_vaccinated_percent
FROM
    `examdataset.examtable`
GROUP BY
    country
ORDER BY
    avg_daily_vaccinations DESC;
"""
]

# Destination bucket details
bucket_name = "ccbd-exam-2024-alghasi"

# Loop through each query and export the results
for idx, query in enumerate(queries):
    # Create a unique temporary table for each query
    temp_table_id = f"exam-dec-2024.examdataset.temp_table_{idx}"

    # Run the query and save the results into the temporary table
    job_config = bigquery.QueryJobConfig(destination=temp_table_id)
    query_job = client.query(query, job_config=job_config)
    query_job.result()  # Wait for the query to complete

    print(f"Query {idx + 1} results saved to temporary table: {temp_table_id}")

    # Configure export job
    destination_uri = f"gs://{bucket_name}/query_results_{idx + 1}_*.csv"
    extract_job_config = bigquery.ExtractJobConfig(
        destination_format=bigquery.DestinationFormat.CSV,
        field_delimiter=",",
        print_header=True
    )

    # Create the extract job to export results to GCS
    extract_job = client.extract_table(
        source=temp_table_id,
        destination_uris=destination_uri,
        job_config=extract_job_config
    )
    extract_job.result()  # Wait for the export job to complete

    print(f"Exported Query {idx + 1} results to {destination_uri}")
```
